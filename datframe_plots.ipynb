# Python Data Frame Visualizations 

#### This Notebook doesn't use any package for Visualizations


```python
import pandas as pd
pd.__version__
```

```python
df_students=pd.read_parquet('students.parquet', engine='pyarrow', dtype_backend='pyarrow')
df_students.rename(columns={'race/ethnicity':'race'},inplace=True)
df_students
```

### Pandas DataFrame has these available options with Plot() :
##### 1.Line
##### 2.Bar
##### 3.Boxplot
##### 4.Pie
##### 5.Kde/Density
##### 6.Hexagonal Bin
##### 7.Histogram
##### 8.Area
##### 9.Scatter


```python
### Line Plot

df_students['math score'].plot(kind='line', title='Line Plot for Math Score')
```

```python
df_students.plot(kind='line', title='Line Plot for Scores')
```

```python
df_students.plot(kind='line', title='Line Plot for Scores', subplots=True)
```

```python
df_students.plot(kind='hist', title='Histogram Plot for Scores', subplots=True)
```

```python
df_students['gender'].value_counts().plot(kind='bar')
```

```python
df_students['gender'].value_counts().plot(kind='barh')
```

```python
df_students['math score'].plot(kind='hist')
```

```python
df_students['writing score'].plot(kind='box')
```

```python
df_students.plot(kind='box')
```

```python
df_students.groupby('parental level of education').plot(kind='box', subplots=True)
```

```python
df_students.plot(x='math score', y='reading score',kind='hexbin')
```

```python
df_student_stats=pd.DataFrame(df_students.describe())
df_student_stats=df_student_stats.loc[['std','mean','50%','75%']]
df_student_stats
df_student_stats.plot.pie(y='math score')
```

```python
df_students.groupby('lunch').size().plot(kind='pie')
```

```python
df_students.groupby('race').size().plot(kind='pie',autopct='%.2f')
```

```python
df_students['math score'].plot(kind='line')
```

```python
df_students['math score'].plot(kind='area')
```

```python
df_students.plot(x='math score', y='reading score', kind='scatter')
```

```python
df_students.groupby('lunch').size().plot(kind='pie')
```

```python
from statsmodels.graphics.mosaicplot import mosaic

mosaic(df_students,['gender','race','lunch'])
```
